```python
from scipy.stats import norm
```

```python
# Эти 4 параметра надо уметь менять.
distribution0 = norm(loc=-1, scale=1)
distribution1 = norm(loc=1, scale=1.2)
```

```python
x = numpy.linspace(-4, 4, num=1000)

figure(figsize=[14, 8])
subplot(121)
plot(x, distribution0.pdf(x), color='r', label='bck')
plot(x, distribution1.pdf(x), color='b', label='sig')
ylabel('distributions')
xlabel('x')

subplot(122)
plot(1 - distribution0.cdf(x), 1- distribution1.cdf(x), color='g')
xlabel('FPR = false positive rate')
ylabel('TPR = true positive rate')
```

```python
# интерактивно надо уметь двигать этот параметр
threshold = 2.
```

```python
figure(figsize=[14, 8])
subplot(121)

plot(x, distribution0.pdf(x), color='r', label='bck')
plot(x, distribution1.pdf(x), color='b', label='sig')

x_left = x[x > threshold]
fill_between(x_left, distribution0.pdf(x_left), alpha=0.3, color='r')
fill_between(x_left, distribution1.pdf(x_left), alpha=0.3, color='b')
# линейка, указывающая threshold
plot([threshold, threshold], [0, ylim()[1]])
ylabel('distributions')
xlabel('x')


subplot(122)
plot(1 - distribution0.cdf(x), 1- distribution1.cdf(x), color='g')

xlabel('FPR = false positive rate')
ylabel('TPR = true positive rate')
# рисуем соответствующую порогу точку
plot([1 - distribution0.cdf(threshold)], [1- distribution1.cdf(threshold)], marker='o', )
```
